```python
import numpy as np
import seaborn as sns
mean_one, sd_one = 5,3
mean_two, sd_two = 10, 2
dist_one = np.random.normal(mean_one, sd_one, 1000)
dist_two = np.random.normal(mean_two, sd_two, 1000)
sns.distplot(dist_one)
sns.distplot(dist_two)
```

```python
sns.distplot([(x - dist_one.mean()) / dist_one.std() for x in dist_one])
sns.distplot([(x - dist_two.mean()) / dist_two.std() for x in dist_two])
```
