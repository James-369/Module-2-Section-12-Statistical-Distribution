```python
size_and_count = {17: 10, 22: 10, 27: 18, 32: 6, 37: 8, 42: 10, 47: 5, 52: 3, 57: 4}
# Determing the total number of class 
sum_class = sum(size_and_count.values())
# Dividing each class size value from the total number of classes
pmf = [round(value / sum_class, 3) for value in size_and_count.values()]
size = list(size_and_count.keys())
pmf, size 
```

```python
# Check to see if the output is 1 
import numpy as np
np.array(pmf).sum()
```

```python
# Find the sum of expected values weighted by thier probabilites 
mu = np.multiply(size, pmf).sum()
mu
# The calculation confirms the expected average class size 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.style.use('ggplot')
plt.figure(figsize = (8,5))
# plot the pmf 
plt.bar(size, pmf)
plt.title("Probability Mass Function")
```

```python
biased = np.multiply(size, pmf)
biased_sum = biased.sum()
biased, biased_sum
```

```python
# Normalize the new biased list with the sum of the listed values 
second_pmf = []
for b in biased:
    second_pmf.append(round(b / biased_sum, 3))
size, second_pmf
```

```python
mu_biased = np.multiply(size, second_pmf).sum()
```

```python
new_figure = plt.figure(figsize = (14, 5.5))
ax = new_figure.add_subplot(121)
ax2 = new_figure.add_subplot(122)
ax.bar(size_and_count.keys(), pmf)
ax2.bar(size, pmf, color = 'yellow')
ax.set_title("Probability Mass Function Actual")
ax2.set_title("Probability Mass Function Observed")
# I observed that there are fewer small classes and more larger classes 
# The mean average of the biased distribution one was higher than the actual mean 

```

```python
# Plotting the pmf's overlapping 
plt.figure(figsize = (8,5))
plt.bar(size, pmf, label = 'Actual', alpha = 0.7)
plt.bar(size, second_pmf, color = 'yellow', label = 'Observed', alpha = 0.7)
plt.title("Probalbility Mass Function - Class Size Paradox")
plt.legend()
plt.show()
```
