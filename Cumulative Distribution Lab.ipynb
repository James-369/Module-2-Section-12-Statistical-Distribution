```python
def calculate_cdf(lst, x):
    count = 0
    for value in lst:
        if value <= x:
            count += 1
    cum_prob = count / len(lst)
    return round(cum_prob , 3)

# Test Data 
test_1st = [1,2,3]
test_x = 2 
calculate_cdf(test_1st, test_x)
```

```python
dice_1st = [1,2,3,4,5,6]
dice_cum = []
for x in dice_1st:
    dice_cum.append(calculate_cdf(test_1st, x))
dice_cum
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
plt.style.use('ggplot')

plt.bar(dice_1st, dice_cum, width = 0.3)
plt.title("Dice Roll - Cumulative Distributiob Function")
plt.xlabel('Dice Values')
plt.ylabel('Cumulative  Probabilites')
```

```python
import numpy as np
sample = [1,2,3,4,5,6]
hist = np.histogram(sample, bins = 6, range = (1,7), normed = True)
plt.scatter(hist[1][:-1], np.cumsum(hist[0]))
plt.title('Dice Roll - Cumulative Distribution Function')
plt.xlabel('Dice Values')
plt.ylabel('Cumulative Probabilites')
plt.yticks(np.linspace(0,1,num = 7))
plt.show()
```
